# <p style="background-color: #000000; font-family:calibri; color: #7FFF00; font-size:140%; text-align:center; border : 4px dashed #7FFF00 ;border-radius:90px 90px 90px 90px;"> Data Normalization Inspection 

This is a Python script that inspects a dataset for normalization issues. It checks for:
1. **Missing Values**: Identifies columns with missing values and suggests imputation methods.
2. **Outliers**: Detects outliers in numerical columns using the Z-score method and suggests handling.
3. **Skewness**: Detects Skewness in features.

Checking features Normality is necesarry before removing outliers if we remove datapoints thinking as outliers they wont give us accurate prediction. So it should be a gebneral practice to normalize the data before checking outliers.

Here we have same features which are already labeld such as:
-  ``Gender``: ``0`` is ``Female`` & ``1`` is ``Male``.
-  ``Smoking``, ``Hypertention``, ``Diabetes`` : ``1`` means they are smoker or having diabetes and hypertion and ``0`` mean not.
-  ``Diagnosis`` : ``1`` means they have liver disease and ``0`` means they don't.
-  ``GeneticRisk`` : ``0`` is Low, ``1`` is Medium, ``2`` is High.

```python
heading("data normality visualization")
sns.set_theme(rc={'axes.facecolor':'#90EE90','figure.facecolor':'#000000'})
plt.figure(figsize=(15,20))

# Selecting Features from original datadet as "Age_group" coliumn was created for easy analysis of data
X = liver.drop(columns=['Diagnosis', 'Age_group'])

# Selecting Label/Target Variable
y = liver['Diagnosis']

for i, col in enumerate(X.columns,1):
    
    plt.subplot(5, 3, i)
    skewness = X[col].skew()
    sns.histplot(x=X[col], kde=True, label = "Skew = %.3f" %(skewness))
    plt.tick_params(axis='both', colors='lightgreen')
    plt.ylabel('Frequency', color='lightgreen')
    plt.title(f'Skewness of {col} Data', fontsize=14,fontweight='bold',color='lightgreen')
    plt.xlabel(X.columns[i-1], fontsize=14,fontweight='bold',color='lightgreen')
    plt.legend()
    plt.grid(False)
    
plt.tight_layout()
```

# <p style="background-color: #000000; font-family:calibri; color: #7FFF00; font-size:140%; text-align:center; border : 4px dashed #7FFF00 ;border-radius:90px 90px 90px 90px;"> Normalizing Features 

We have used ``Quantile Transformer`` to reduce the skewness in features and normalized the data.
Skewness is also show as comparison before and after the normalization.

```python
sns.set_theme(rc={'axes.facecolor':'#90EE90','figure.facecolor':'#000000'})
plt.figure(figsize=(15,20))

transformed_features = QuantileTransformer(output_distribution='normal')
transformed_df = transformed_features.fit_transform(X)
normalized_features = pd.DataFrame(transformed_df, columns=X.columns)

normalized_features.head(3)

for i, col in enumerate(normalized_features.columns,1):
    
    plt.subplot(5, 3, i)
    skewness = normalized_features[col].skew()
    sns.histplot(x=normalized_features[col], kde=True, label = "Skew = %.3f" %(skewness))
    plt.tick_params(axis='both', colors='lightgreen')
    plt.ylabel('Frequency', color='lightgreen')
    plt.title(f'Skewness of {col} Data', fontsize=14,fontweight='bold',color='lightgreen')
    plt.xlabel(normalized_features.columns[i-1], fontsize=14,fontweight='bold',color='lightgreen')
    plt.legend()
    plt.grid(False)
    
plt.tight_layout()
```

# <p style="background-color: #000000; font-family:calibri; color: #7FFF00; font-size:140%; text-align:center; border : 4px dashed #7FFF00 ;border-radius:90px 90px 90px 90px;"> Splitting Data into Test and Train Sets


```python
X_train, X_test, y_train, y_test = train_test_split(normalized_features, y, train_size=0.7, random_state=42)
```

<a id="Import"></a>
# <p style="background-color: #000000; font-family:calibri; color: #7FFF00; font-size:140%; text-align:center; border : 4px dashed #7FFF00 ;border-radius:90px 90px 90px 90px;"> Instantiating Classification Models

Evaluation of CV_Scores for all models then taking mean CV Score and displaying on bars.

```python
models = [LogisticRegression(), KNeighborsClassifier(), DecisionTreeClassifier(), SVC(), RandomForestClassifier(), GradientBoostingClassifier(), GaussianNB()]
names = ['LogReg','KNN', 'DTClassifier','SVC', 'RFClassifier','GBClassifier', 'GaussianNB']   
results=[] 

for name, model in zip(names, models):
    kf=KFold(n_splits=4, shuffle=True, random_state=42)
    cv_results = cross_val_score(model, X_train, y_train, cv=kf)
    average_results = round(cv_results.mean(), 3)
    results.append([name, average_results])
    
sorted_models = sorted(results, key=lambda x: x[1], reverse=True)
df = pd.DataFrame(sorted_models, columns=['Model', 'CV_Score'])
heading('Sorted CV_Scores')
print(df)
heading('Barplot of CSV_Scores')      

fig = px.bar(df, x='Model', y='CV_Score', text_auto=True)
fig.update_layout(title = 'Cross Validation Scores', title_x=0.5, font_color='darkblue', font_family='Arial Black', height=600, width=700, plot_bgcolor='lightgrey', font_size=15)
fig.update_xaxes(title='Classification Models', tickfont_family="Arial Black", linecolor='black', showgrid=False)
fig.update_yaxes(title='Cross Val Score', tickfont_family="Arial Black", linecolor='black', showgrid=False)
fig.update_traces(textposition='outside', marker_line_width=2, opacity=0.6, textfont_family='Arial Black')
fig.show()
```

# <p style="background-color: #000000; font-family:calibri; color: #7FFF00; font-size:140%; text-align:center; border : 4px dashed #7FFF00 ;border-radius:90px 90px 90px 90px;"> Hyperparameter Tuning  and Classification Reports 

we have tunes hyperparameters of various classification models and printed classification report of each models which gives ``Precision``, ``Accuracy``, ``Recall`` and ``F1 Score`` for each positive and negative class.

```python
heading('Models Tunes Parameters')
models = {'LogReg':LogisticRegression(), 'KNN':KNeighborsClassifier(), 'DTC': DecisionTreeClassifier(), 'SVC' :SVC(), 
          'RandomForest':RandomForestClassifier(), 'GradientBoosting': GradientBoostingClassifier(),'Naive Bayes':GaussianNB(),
          'VotingClassifier':VotingClassifier(estimators=[('LogReg',LogisticRegression()), ('KNN',KNeighborsClassifier()),
                                                          ('DTC', DecisionTreeClassifier()), ('SVC' ,SVC()), ('RandomForest',RandomForestClassifier()),
                                                          ('GradientBoosting', GradientBoostingClassifier()),
                                                          ('Naive Bayes',GaussianNB())], voting='hard')}
param_grids = {'LogReg' : {'C': [0.1,1,10,100], 'solver':['lbfgs','liblinear'], 'penalty':['l2']},
                'KNN'   : {'n_neighbors': [3, 5, 7, 9], 'weights':['uniform','distance'], 'metric':['euclidean', 'manhattan']},
                'DTC'   : {'max_depth': [None, 10, 20, 30, 40], 'min_samples_split': [2, 5, 10], 'criterion':['gini', 'entropy', 'log_loss']},
                'SVC'   : {'kernel': ['poly', 'rbf'], 'gamma':[1, 0.1, 0.01, 0.001], 'C': [0.1, 0.4, 0.1]},
                'XGBoost' : {'max_depth': [3,5,7],'n_estimators':[50, 60, 70], 'subsample':[0.6, 0.8, 1], 'learning_rate':[0.01, 0.1, 0.2]},
                'RandomForest' :{'n_estimators':[10, 20, 30, 40], 'max_depth': [None, 10, 20, 30, 40], 'min_samples_split': [2, 5, 10]},
                'GradientBoosting' : {'n_estimators':[10, 20, 30], 'learning_rate':[0.01, 0.1, 0.2], 'max_depth': [3,5,7]},
                'Naive Bayes' : {'var_smoothing': np.logspace(0,-9, num=100)}, 
                'VotingClassifier': {}}
scores=[]  

for model_name, classifier in models.items():
    rcv = RandomizedSearchCV(estimator=models[model_name], param_distributions=param_grids[model_name], cv=3, scoring='accuracy', verbose=1, n_jobs=-1, n_iter=2, random_state=42)
    rcv.fit(X_train,y_train)
    y_pred = rcv.predict(X_test)
    print(f'\033[1m{model_name}\033[0m best parameters are : \033[1m{rcv.best_params_}\033[0m ')
    print(f'\033[1m{model_name}\033[0m Classification_Report is : \n \033[1m{classification_report(y_test, y_pred)}\033[0m  ')
    print('\n')
```

# <p style="background-color: #000000; font-family:calibri; color: #7FFF00; font-size:140%; text-align:center; border : 4px dashed #7FFF00 ;border-radius:90px 90px 90px 90px;"> Confusion Matrix Calculations and Visualization

This repository provides a Python script to calculate and visualize a confusion matrix for a classification problem.

```python
heading('Confusion Matrix')
rfc = RandomForestClassifier(n_estimators=40, min_samples_split=2, max_depth=20)
rfc.fit(X_train, y_train)
train_score=rfc.score(X_train, y_train)
print(train_score)
y_pred = rfc.predict(X_test)
test_score=rfc.score(X_test, y_test)
print(test_score)
cm = confusion_matrix(y_test, y_pred)
print(cm)
fig, ax = plt.subplots()
dis = ConfusionMatrixDisplay(confusion_matrix=cm)
dis.plot(cmap='viridis', values_format='d', ax=ax, colorbar=False)

plt.xlabel('Predicted', fontsize=14,fontweight='bold', color='lightgreen')
plt.ylabel('Actual', fontsize=14,fontweight='bold', color='lightgreen')
plt.yticks([0,1],['Not-Diagnosed','Diagnosed'], color='lightgreen')
plt.xticks([0,1],['Not-Diagnosed','Diagnosed'],color='lightgreen')
plt.xticks(fontsize=12,fontweight='bold', color='lightgreen')
plt.yticks(fontsize=12,fontweight='bold', color='lightgreen')
plt.title("Confusion Matrix", fontsize=14,fontweight='bold',color='lightgreen')
ax.grid(False)
```
